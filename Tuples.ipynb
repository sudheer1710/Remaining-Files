```python
fhand = open('polarity_pos.txt')
counts = dict()
for line in fhand:
    words = line.split()
    for word in words:
        counts[word] = counts.get(word, 0) + 1
        
lst = list()
for key,val in counts.items():
    lst.append((val,key))
    
lst.sort(reverse=True)

for var, key in lst[:10]:
    print(key,val)
```

```python
fhand = open('polarity_pos.txt')
counts = dict()
for line in fhand:
    words = line.split()
    for word in words:
        counts[word] = counts.get(word, 0) + 1
        
print(sorted((v,k) for k,v in counts.items()))
```

```python
# Shorter version

c = {'a': 10, 'b':1, 'c':22}

print(sorted((v,k) for k,v in c.items()))
```
