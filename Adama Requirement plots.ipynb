```python
plt.figure(figsize=(20,10), dpi=1200)
p1 = sns.scatterplot('Market..INR.Cr.', # Horizontal axis
       'ADAMA..INR.Cr.', # Vertical axis
       data=df3, # Data source
       size = 8,
       legend=False)  

for line in range(0,df3.shape[0]):
     p1.text(df3['Market..INR.Cr.'][line]+0.01, df3['ADAMA..INR.Cr.'][line], 
     df3.Territory[line], horizontalalignment='left', 
     size='medium', color='black', weight='semibold')

plt.title('Scatterplot')
# Set x-axis label
plt.xlabel('Market..INR.Cr.')
# Set y-axis label
plt.ylabel('ADAMA..INR.Cr.')
```
