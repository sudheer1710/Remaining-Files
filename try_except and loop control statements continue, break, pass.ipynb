```python
import sys
```

```python
randomlist=['a',0,2]
for entry in randomlist:
    try:
        print("The entry is",entry)
        r=1/int(entry)
        break                                        
    except:
        print("Oops!",sys.exc_info()[0],"occured")
        print("Next Entry")
        print()
print("The reciprocal of",entry,"is",r)
```

```python
for i in range(1,101):
    print(i)
    if i>5:
        continue                                  # continue statement breaks off wherever it is and starts for loop again.
    else:
        print('else statement printing')
    print('loop statement printing')
```
